```python
class Dataset(object):  # lớp Dataset
    def __getitem__(self, index): # hàm lấy một mẫu từ dữ liệu
        raise NotImplementedError # ném một ngoại lệ nếu không được triển khai

    def __len__(self): # hàm trả về độ dài của dữ liệu
        raise NotImplementedError # ném một ngoại lệ nếu không được triển khai

    def __add__(self, other): # hàm nối hai tập dữ liệu
        return ConcatDataset([self, other]) # nối hai tập dữ liệu
```

```python
class MRI(Dataset): # lớp MRI kế thừa từ lớp Dataset
    def __init__(self): # hàm khởi tạo

        tumor = [] # mảng để lưu các hình ảnh của lớp tumor
        strong = [] # mảng để lưu các hình ảnh của lớp strong
        # cv2 - It reads in BGR format by default
        for f in glob.iglob("drive/MyDrive/brain_tumor_dataset/yes/*jpg"):
            img = cv2.imread(f)
            img = cv2.resize(img,(128,128))
            b, g, r = cv2.split(img)
            img = cv2.merge([r,g,b])
            img = img.reshape((img.shape[2],img.shape[0],img.shape[1])) # cách chuyển đổi hình ảnh thành kích thước (c,h,w), nếu không sẽ là (h,w,c)
            tumor.append(img) # thêm hình ảnh vào mảng tumor

        for f in glob.iglob("drive/MyDrive/brain_tumor_dataset/no/*jpg"):
            img = cv2.imread(f)
            img = cv2.resize(img,(128,128))
            b, g, r = cv2.split(img)
            img = cv2.merge([r,g,b])
            img = img.reshape((img.shape[2],img.shape[0],img.shape[1]))
            strong.append(img)

        # our images
        tumor = np.array(tumor,dtype=np.float32) # chuyển mảng tumor thành mảng numpy
        strong = np.array(strong,dtype=np.float32) # chuyển mảng strong thành mảng numpy

        # our labels
        tumor_label = np.ones(tumor.shape[0], dtype=np.float32) # tạo nhãn cho lớp tumor, 1 là lớp tumor
        strong_label = np.zeros(strong.shape[0], dtype=np.float32) # tạo nhãn cho lớp strong, 0 là lớp strong

        # Concatenates
        self.images = np.concatenate((tumor, strong), axis=0) # nối các hình ảnh của lớp tumor và lớp strong, axis=0 nối theo chiều dọc, axis=1 nối theo chiều ngang, axis=2 nối theo chiều sâu,
        self.labels = np.concatenate((tumor_label, strong_label)) # nối các nhãn của lớp tumor và lớp strong,

    def __len__(self): # hàm trả về độ dài của dữ liệu
        return self.images.shape[0] # trả về số lượng hình ảnh trong dữ liệu

    def __getitem__(self, index): # hàm lấy một mẫu từ dữ liệu ,

        sample = {'image': self.images[index], 'label':self.labels[index]} # lấy một mẫu từ dữ liệu

        return sample # trả về mẫu

    def normalize(self): # hàm chuẩn hóa dữ liệu
        self.images = self.images/255.0 # chuẩn hóa dữ liệu ,
```

```python
import torch.nn as nn # thư viện torch.nn để tạo mô hình
import torch.nn.functional as F # thư viện torch.nn.functional để sử dụng các hàm kích hoạt

class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.cnn_model = nn.Sequential(
            nn.Conv2d(in_channels=3, out_channels=64, kernel_size=5),
            nn.BatchNorm2d(64),
            nn.Tanh(),
            nn.AvgPool2d(kernel_size=2, stride=2),
            nn.Conv2d(in_channels=64, out_channels=32, kernel_size=5),
            nn.BatchNorm2d(32),
            nn.Tanh(),
            nn.AvgPool2d(kernel_size=2, stride=2)
        )
        dummy_input = torch.randn(1, 3, 128, 128)
        dummy_output = self.cnn_model(dummy_input)
        conv_output_size = dummy_output.numel() // dummy_output.size(0)

        self.fc_model = nn.Sequential(
            nn.Linear(in_features=conv_output_size, out_features=150),
            nn.InstanceNorm1d(150),
            nn.Tanh(),
            nn.Dropout(0.5),
            nn.Linear(in_features=150, out_features=100),
            nn.InstanceNorm1d(100),
            nn.Tanh(),
            nn.Dropout(0.5),
            nn.Linear(in_features=100, out_features=1)
        )
    def forward(self, x):
        x = self.cnn_model(x)
        x = x.view(x.size(0), -1)
        x = self.fc_model(x)  # Pass through fully connected layers
        x = torch.sigmoid(x)
        return x
```

```python
def threshold(scores,threshold=0.50, minimum=0, maximum = 1.0): # hàm để chuyển đổi giá trị dự đoán thành các lớp
    x = np.array(list(scores)) # chuyển scores thành mảng numpy, x sẽ chứa giá trị dự đoán, threshold là ngưỡng
    x[x >= threshold] = maximum
    x[x < threshold] = minimum
    return x
accuracy_score(y_true, threshold(outputs))
```

```python
eta = 0.0001 # tốc độ học
EPOCH = 400 # số lượng vòng lặp
optimizer = torch.optim.Adam(model.parameters(), lr=eta) # tối ưu hóa mô hình với Adam  và tốc độ học eta
dataloader = DataLoader(mri_dataset, batch_size=32, shuffle=True) # tạo một dataloader với kích thước batch là 32 và xáo trộn dữ liệu
model.train() # chuyển mô hình sang chế độ huấn luyện
```

```python
for epoch in range(1, EPOCH): # vòng lặp qua số lượng epoch đã chọn
    losses = [] # mảng để lưu giá trị hàm mất mát
    for D in dataloader: # vòng lặp qua các batch dữ liệu
        optimizer.zero_grad() # đặt đạo hàm của các tham số về 0
        data = D['image'].to(device) # chuyển dữ liệu vào thiết bị
        label = D['label'].to(device) # chuyển nhãn vào thiết bị
        y_hat = model(data) # dự đoán với mô hình
        # define loss function
        error = nn.BCELoss() # hàm mất mát là hàm mất mát nhị phân chéo
        loss = torch.sum(error(y_hat.squeeze(), label)) # tính toán hàm mất mát
        loss.backward() # lan truyền ngược để tính toán đạo hàm
        optimizer.step() # cập nhật tham số
        losses.append(loss.item())  # lưu giá trị hàm mất mát
    if (epoch+1) % 10 == 0: # in ra giá trị hàm mất mát sau mỗi 10 epoch
        print('Train Epoch: {}\tLoss: {:.6f}'.format(epoch+1, np.mean(losses))) # in ra giá trị hàm mất mát trung bình
```

```python
from sklearn.model_selection import train_test_split # hàm chia dữ liệu thành tập huấn luyện và tập kiểm tra

class MRI(Dataset): # lớp MRI kế thừa từ lớp Dataset

    def __init__(self): # hàm khởi tạo

        # Variables to hold the Training data and Validation data
        self.X_train, self.y_train, self.X_val, self.y_val = None, None, None, None  # biến để lưu dữ liệu huấn luyện và dữ liệu xác thực


        # 1 biến để xác định xem chúng ta quan tâm đến việc lấy dữ liệu huấn luyện HOẶC dữ liệu xác thực
        # A variable to determine if we are interested in retrieving the training OR the validation data
        self.mode = 'train' # chế độ mặc định là huấn luyện

        tumor = [] # mảng để lưu các hình ảnh của lớp tumor
        strong = [] # mảng để lưu các hình ảnh của lớp strong
        # cv2 - It reads in BGR format by default
        for f in glob.iglob("drive/MyDrive/brain_tumor_dataset/yes/*.jpg"):
            img = cv2.imread(f)
            img = cv2.resize(img,(128,128)) # I can add this later in the boot-camp for more adventure
            b, g, r = cv2.split(img)
            img = cv2.merge([r,g,b])
            img = img.reshape((img.shape[2],img.shape[0],img.shape[1])) # otherwise the shape will be (h,w,#channels)
            tumor.append(img)

        for f in glob.iglob("drive/MyDrive/brain_tumor_dataset/no/*.jpg"):
            img = cv2.imread(f)
            img = cv2.resize(img,(128,128))
            b, g, r = cv2.split(img)
            img = cv2.merge([r,g,b])
            img = img.reshape((img.shape[2],img.shape[0],img.shape[1]))
            strong.append(img)

        # our images
        tumor = np.array(tumor,dtype=np.float32) # chuyển mảng tumor thành mảng numpy
        strong = np.array(strong,dtype=np.float32) # chuyển mảng strong thành mảng numpy

        # our labels
        tumor_label = np.ones(tumor.shape[0], dtype=np.float32) # tạo nhãn cho lớp tumor
        healthy_label = np.zeros(strong.shape[0], dtype=np.float32) # tạo nhãn cho lớp strong

        # Concatenates
        self.images = np.concatenate((tumor, strong), axis=0) # nối các hình ảnh của lớp tumor và lớp strong
        self.labels = np.concatenate((tumor_label, healthy_label)) # nối các nhãn của lớp tumor và lớp strong

    # Define a function that would separate the data into Training and Validation sets
    def train_val_split(self): # hàm chia dữ liệu thành tập huấn luyện và tập xác thực
        self.X_train, self.X_val, self.y_train, self.y_val = train_test_split(self.images, self.labels, test_size=0.20, random_state=42)  # chia dữ liệu thành 80% huấn luyện và 20% xác thực

    def __len__(self): # hàm trả về độ dài của dữ liệu
        # Use self.mode to deetrmine whether train or val data is of interest
        if self.mode == 'train': # nếu chế độ là huấn luyện
            return self.X_train.shape[0] # trả về số lượng hình ảnh trong tập huấn luyện
        elif self.mode == 'val': # nếu chế độ là xác thực
            return self.X_val.shape[0] # trả về số lượng hình ảnh trong tập xác thực

    def __getitem__(self, idx): # hàm lấy một mẫu từ dữ liệu
        # Use self.mode to deetrmine whether train or val data is of interest
        if self.mode== 'train': # nếu chế độ là huấn luyện
            sample = {'image': self.X_train[idx], 'label': self.y_train[idx]} # lấy một mẫu từ tập huấn luyện

        elif self.mode== 'val': # nếu chế độ là xác thực
            sample = {'image': self.X_val[idx], 'label': self.y_val[idx]} # lấy một mẫu từ tập xác thực

        return sample # trả về mẫu

    def normalize(self): # hàm chuẩn hóa dữ liệu
        self.images = self.images/255.0 # chuẩn hóa dữ liệu
```

```python
train_dataloader = DataLoader(mri_dataset, batch_size=32, shuffle=True) # tạo một dataloader cho tập huấn luyện
val_dataloader = DataLoader(mri_dataset, batch_size=32, shuffle=False) # tạo một dataloader cho tập xác thực

device = torch.device("cuda:0") # thiết bị sẽ là 'cuda' nếu có GPU
model = CNN().to(device)
```

```python
eta=0.0001 # tốc độ học
optimizer = torch.optim.Adam(model.parameters(), lr=eta) # tối ưu hóa mô hình với Adam và tốc độ học eta

# keep track of epoch losses
epoch_train_loss = [] # mảng để lưu giá trị hàm mất mát của tập huấn luyện
epoch_val_loss = [] # mảng để lưu giá trị hàm mất mát của tập xác thực
```

```python
for epoch in range(1,600): # vòng lặp qua số lượng epoch đã chọn
    train_losses = [] # mảng để lưu giá trị hàm mất mát của tập huấn luyện
    val_losses = []
    # train for the current epoch
    model.train() # chuyển mô hình sang chế độ huấn luyện
    mri_dataset.mode = 'train' # chuyển chế độ của tập dữ liệu thành huấn luyện
    mri_dataset.mode = 'val'
    for D in train_dataloader: # vòng lặp qua các batch dữ liệu
        # Train the model
        optimizer.zero_grad() # đặt đạo hàm của các tham số về 0
        data = D['image'].to(device) # chuyển dữ liệu vào thiết bị
        label = D['label'].to(device) # chuyển nhãn vào thiết bị

        y_hat = model(data) # dự đoán với mô hình
        error = nn.BCELoss() # hàm mất mát là hàm mất mát nhị phân chéo
        loss = torch.sum(error(y_hat.squeeze(), label)) # tính toán hàm mất mát
        loss.backward() # lan truyền ngược để tính toán đạo hàm
        optimizer.step() # cập nhật tham số
        train_losses.append(loss.item()) # lưu giá trị hàm mất mát
    epoch_train_loss.append(np.mean(train_losses))
    model.eval() # chuyển mô hình sang chế độ đánh giá
    with torch.no_grad():
        for D in val_dataloader:
            data = D['image'].to(device)
            label = D['label'].to(device)
            y_hat = model(data)
            error = nn.BCELoss()
            loss = torch.sum(error(y_hat.squeeze(), label))
            val_losses.append(loss.item())
        epoch_val_loss.append(np.mean(val_losses))
    if (epoch+1) % 10 == 0: # in ra giá trị hàm mất mát sau mỗi 10 epoch
            print(len(epoch_train_loss))
            print('Train Epoch: {}\tTrain Loss: {:.6f}\tVal Loss: {:.6f}'.format(epoch+1, np.mean(train_losses),np.mean(val_losses))) # in ra giá trị hàm mất mát trung bình

    epoch_train_loss.append(np.mean(train_losses))
```

```python
plt.figure(figsize=(16,9)) # kích thước của hình ảnh
plt.plot(epoch_train_loss, c='b', label='Train loss') # vẽ đồ thị hàm mất mát của tập huấn luyện
plt.plot(epoch_val_loss, c='r', label = 'Validation loss') # vẽ đồ thị hàm mất mát của tập xác thực
plt.legend() # thêm chú thích
plt.grid() # vẽ lưới
plt.xlabel('Epochs', fontsize=20) # đặt nhãn cho trục x
plt.ylabel('Loss', fontsize=20) # đặt nhãn cho trục y
```
